5- Calidad de Datos: Identificar valores nulos y duplicados en los conjuntos de datos. Documentar el estado inicial de los datos.

```python
#Identificamos duplicados en cada DATASETS, considerando que el primer valor es correcto y los que se repiten, son sus duplicados.
print("--------------------- Duplicados de cada DATASETS -----------------------------")
print("Ventas: ")
print("----------------------------------------------------------------------------------")
print(f"Cantdidad de filas duplicadas: {df_ventas.duplicated(keep=False).sum()}")
print("----------------------------------------------------------------------------------")
print("Clientes: ")
print("----------------------------------------------------------------------------------")
print(f"Cantidad de filas duplicadas: {df_clientes.duplicated().sum()}")
print("----------------------------------------------------------------------------------")
print("Marketing: ")
print("----------------------------------------------------------------------------------")
print(f"Cantidad de filas duplicadas: {df_marketing.duplicated().sum()}")
```

```python
#DEF para mostrar contar los duplicados y mostrar los datos filtrados:
def calidad(df,nombre,id):
  print("Valores duplicados de la table",nombre)
  print("-")
  print(f"Cantidad de filas duplicadas: {df.duplicated(keep=False).sum()}")
  print("-")
  #contar duplicados por id de cada df
  print(f"Cantidad de filas duplicadas por {id}: {df.duplicated(keep=False).sum()}")
  print("-")
  #Armar una tabla filtrando los valores duplicados para mostrarlos
  duplicados = (df[df[id].duplicated(keep=False)].sort_values(by=id))
  print(duplicados.head(10))
  print("-")
```

```python
#Ejecucion de nuestro DEF
calidad(df_ventas, "Ventas", "id_venta")
calidad(df_clientes, "Clientes", "id_cliente")
calidad(df_marketing, "Marketing", "id_campanha")
```

6 - Limpieza de datos

```python
#Copia de cada DataFrame
df_ventas_clean = df_ventas.copy()
df_clientes_clean = df_clientes.copy()
df_marketing_clean = df_marketing.copy()
#eliminar duplicados de cada DF
df_ventas_clean= df_ventas_clean.drop_duplicates()
df_clientes_clean= df_clientes_clean.drop_duplicates()
df_marketing_clean= df_marketing_clean.drop_duplicates()
print("Información de df_ventas_clean después de eliminar duplicados:")
df_ventas_clean.info()
print("\nInformación de df_clientes_clean después de eliminar duplicados:")
df_clientes_clean.info()
print("\nInformación de df_marketing_clean después de eliminar duplicados:")
df_marketing_clean.info()
print("\nCantidad de duplicados en df_ventas_clean:")
print(df_ventas_clean.duplicated().sum())
```

```python

#buscar columnas precios / costo / ingresos y eliminar signo $  - espacios
df_ventas_clean['precio'] = df_ventas_clean['precio'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).str.strip()
df_clientes_clean['ingresos'] = df_clientes_clean['ingresos'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).str.strip()
df_marketing_clean['costo'] = df_marketing_clean['costo'].astype(str).str.replace('$', '', regex=False).str.replace(',', '', regex=False).str.strip()
print("datos de la columna precio: ")
df_ventas_clean.info()
print("----------------------------------------------------------------------------------")
print(df_ventas_clean['precio'].head(5).sort_values(ascending=True))
print(df_clientes_clean['ingresos'].head(5).sort_values(ascending=True))
print(df_marketing_clean['costo'].head(5).sort_values(ascending=True))
#
#
```

```python
#Transformamos las columnas precio, ingresos y costo de nuestro dataframe en columna precio númerico.
df_ventas_clean['precio'] = pd.to_numeric(df_ventas_clean['precio'], errors='coerce')
df_clientes_clean['ingresos'] = pd.to_numeric(df_clientes_clean['ingresos'], errors='coerce')
df_marketing_clean['costo'] = pd.to_numeric(df_marketing_clean['costo'], errors='coerce')
```

```python
#Limpiar texto de columnas tipo string
#borramos los espacios - convertimos a tipo string - convertimos las expresiones regulares - reempleazamos los espacios y convertimos los nombres propios con mayusculas las iniciales.
#Ventas
for col in df_ventas_clean.select_dtypes(include=['object']).columns:
  df_ventas_clean[col] = df_ventas_clean[col].astype(str).str.strip().replace(r"[\u200b\t\r\n]", "", regex=True).str.replace(" +", " ", regex=True).str.title()
#Clientes
for col in df_clientes_clean.select_dtypes(include=['object']).columns:
  df_clientes_clean[col] = df_clientes_clean[col].astype(str).str.strip().replace(r"[\u200b\t\r\n]", "", regex=True).str.replace(" +", " ", regex=True).str.title()
#Marketing
for col in df_marketing_clean.select_dtypes(include=['object']).columns:
  df_marketing_clean[col] = df_marketing_clean[col].astype(str).str.strip().replace(r"[\u200b\t\r\n]", "", regex=True).str.replace(" +", " ", regex=True).str.title()
```

```python
print(df_ventas_clean.info())
print(df_ventas_clean.head())
print(df_clientes_clean.info())
print(df_clientes_clean.head())
print(df_marketing_clean.info())
print(df_marketing_clean.head())
```

```python
#Transformar y normalizar las columnas fechas de cada dataframe
for df in [df_ventas_clean,df_clientes_clean,df_marketing_clean]:
  for col in df.columns:
    if 'fecha' in col.lower():
      df[col] = pd.to_datetime(df[col], errors='coerce',dayfirst=True)

```
